# High Level Texting Features

```python
# Find the length of each text
tft["cnt"]=tft['text_clean'].str.len()

# Count the number of words in each text
tft["word_cnt"]=tft["text_clean"].str.split(" ").str.len()

# Find the average length of word
tft["Average_word_cnt"]=tft["cnt"]/tft["word_cnt"]
```

# Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

count=CountVectorizer()

count.fit(tft['text_clean'])

count.get_feature_names()

count_tranform=count.transform(tft['text_clean'])
# Need to convert into Array
cv_array=count_tranform.toarray()
cv_array
```

# limiting The Feature

```python
from sklearn.feature_extraction.text import CountVectorizer

limt_cv=CountVectorizer(min_df=0.2,max_df=0.8)
limt_cv.fit(tft["text_clean"])
cv_transformed = limt_cv.fit_transform(tft['text_clean'])
cv_array = cv_transformed.toarray()
cv_array.shape
```

# Now Text to dataFrame

```python
cv_df=pd.DataFrame(cv_array,columns=limt_cv.get_feature_names()).add_prefix("Counts_")
```

```python
cv_df.head()
```

# Tf-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer 

tv=TfidfVectorizer()

tv.fit(tft["text_clean"])

tv_transformed = tv.fit_transform(tft['text_clean'])

tv_df = pd.DataFrame(tv_transformed.toarray(), 
                     columns=tv.get_feature_names()).add_prefix("Tfidf_")
```

```python
tv_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv_tfidf=TfidfVectorizer(max_features=100,stop_words="english")

transform=tv_tfidf.fit_transform(tft['text_clean'])

tv_vectorizer=pd.DataFrame(transform.toarray(),columns=tv_tfidf.get_feature_names()).add_prefix("tfidf_")
```
